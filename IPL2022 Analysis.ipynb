#### Import the libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

#### Read the dataset

```python
df = pd.read_csv("IPL_Dataset.csv")
```

#### plot a graph for the number of matches won by each team in IPL 2022

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=df["match_winner"].value_counts().index, y=df["match_winner"].value_counts().values, palette="Set2")
plt.title("Number of Matches Won in IPL 2022")
plt.xticks(rotation=90)
plt.show()
```

#### now analyze whether most of the teams win by defending (batting first) or chasing (batting second)

```python
df["won_by"] = df["won_by"].map({"Wickets": "Chasing", "Runs": "Defending"})
plt.figure(figsize=(8, 8))
sns.countplot(x="won_by", data=df, palette=["purple", "lightblue"])
plt.title("Number of Matches Won By Defending Or Chasing")
plt.show()
```

### lets see what most teams prefer (batting or fielding) after winning the toss

```python
plt.figure(figsize=(8, 8))
sns.countplot(x="toss_decision", data=df, palette=["lightblue", "purple"])
plt.title("Toss Decision")
plt.show()
```

### now let’s see the top scorers of most IPL 2022 matches

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=df["top_scorer"].value_counts().index, y=df["top_scorer"].value_counts().values, palette="Set2")
plt.title("Top Scorers in IPL 2022")
plt.xticks(rotation=90)
plt.show()
```

### runs scored by the top scorers

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=df["top_scorer"], y=df["highscore"], hue=df["highscore"], palette="viridis")
plt.title("Top Scorers in IPL 2022 with High Scores")
plt.xticks(rotation=90)
plt.show()
```

### most player of the match awards till now in IPL 2022

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=df["player_of_the_match"].value_counts().index, y=df["player_of_the_match"].value_counts().values, palette="Set2")
plt.title("Most Player of the Match Awards")
plt.xticks(rotation=90)
plt.show()
```

### analysing the bowlers with the best bowling figures in most of the matches

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=df["best_bowling"].value_counts().index, y=df["best_bowling"].value_counts().values, palette="Set2")
plt.title("Best Bowlers in IPL 2022")
plt.xticks(rotation=90)
plt.show()
```

### Now analyze whether most of the wickets fall while setting the target or while chasing the target

```python
venue_data = df.groupby('venue')[['first_ings_wkts', 'second_ings_wkts']].sum().reset_index()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x="venue", y="first_ings_wkts", data=venue_data, color="purple", label="First Innings Wickets")
sns.barplot(x="venue", y="second_ings_wkts", data=venue_data, color="lightblue", label="Second Innings Wickets")
plt.title("Wickets in Different Venues")
plt.xticks(rotation=90)
plt.legend()
plt.show()
```

```python
venue_avg_scores = df.groupby('venue')['first_ings_score'].mean().reset_index()
venue_avg_scores = venue_avg_scores.sort_values(by='first_ings_score', ascending=False)

plt.figure(figsize=(12, 6))
sns.barplot(x=venue_avg_scores['venue'], y=venue_avg_scores['first_ings_score'], palette="viridis")
plt.title("Average Scores at Different Venues")
plt.xticks(rotation=90)
plt.show()
```
